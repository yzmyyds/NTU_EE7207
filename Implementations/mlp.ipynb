```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=pd.read_csv('dataset/unlinear_data.csv').values
y=pd.read_csv('dataset/unlinear_label.csv').values
# X: (500,2), y: (500,1) {-1,1}
```

```python
# mlp configuration
## input layer: 2 neurons
## hidden layer: k neurons
## output layer: 1 neuron
## activation function: sigmoid
X=np.concatenate((X,np.ones((X.shape[0],1))),axis=1) # add bias term

def sigmoid(x):
    return 1/(1+np.exp(-x))

def config(X,y,k):
    m,n=X.shape
    W1=np.random.randn(n,k)
    W2=np.random.randn(k+1,1)
    return W1,W2

def forward(X,W1,W2):
    m,n=X.shape
    z1=X@W1
    a1=sigmoid(z1)
    a1=np.concatenate((a1,np.ones((a1.shape[0],1))),axis=1) # add bias term
    z2=a1@W2
    a2=sigmoid(z2)
    return a1,a2

def backward(X,y,a1,a2,W1,W2):
    m,n=X.shape
    dz2=a2-y
    dw2=a1.T@dz2/m
    dz1=dz2@W2.T*a1*(1-a1)
    dw1=X.T@dz1/m
    return dw1,dw2

def train(X,y,k,eta,max_iter):
    W1,W2=config(X,y,k)
    for i in range(max_iter):
        a1,a2=forward(X,W1,W2)
        dw1,dw2=backward(X,y,a1,a2,W1,W2)
        W1-=eta*dw1
        W2-=eta*dw2
    return W1,W2

def predict(X,W1,W2):
    _,a2=forward(X,W1,W2)
    return np.where(a2>=0.5,1,-1)


def main():
    k=6
    eta=0.1
    max_iter=1000
    W1,W2=train(X,y,k,eta,max_iter)
    y_pred=predict(X,W1,W2)
    print("Accuracy:",np.mean(y_pred==y.flatten()))

if __name__=="__main__":
    main()
    

```
