```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=pd.read_csv('dataset/unlinear_data.csv').values
y=pd.read_csv('dataset/unlinear_label_zp.csv').values
# X: (500,2), y: (500,1) {-1,1}
print(y)
```

```python
# mlp configuration
## input layer: 2 neurons
## hidden layer: k neurons
## output layer: 1 neuron
## activation function: sigmoid

def sigmoid(x):
    return 1/(1+np.exp(-x))

def config(X, y, k):
    m, n = X.shape
    W1 = np.random.randn(n, k)
    b1 = np.zeros((1, k)) 
    
    W2 = np.random.randn(k, 1)
    b2 = np.zeros((1, 1))
    
    return W1, b1, W2, b2

def forward(X,W1,b1,W2,b2):
    m,n=X.shape
    z1=X@W1+b1
    a1=sigmoid(z1)
    z2=a1@W2+b2
    a2=sigmoid(z2)
    return a1,a2

def backward(X,y,a1,a2,W1,W2,b1,b2):
    m,n=X.shape
    dz2=a2-y
    dw2=a1.T@dz2/m
    db2=np.sum(dz2,axis=0,keepdims=True)/m
    dz1=dz2@W2.T*a1*(1-a1)
    dw1=X.T@dz1/m
    db1=np.sum(dz1,axis=0,keepdims=True)/m
    return dw1,db1,dw2,db2

def train(X,y,k,eta,max_iter):
    W1,b1,W2,b2=config(X,y,k)
    for i in range(max_iter):
        a1,a2=forward(X,W1,b1,W2,b2)
        dw1,db1,dw2,db2=backward(X,y,a1,a2,W1,W2,b1,b2)
        W1-=eta*dw1
        b1-=eta*db1
        W2-=eta*dw2
        b2-=eta*db2
    return W1,b1,W2,b2

def predict(X,W1,b1,W2,b2):
    _,a2=forward(X,W1,b1,W2,b2)
    return np.where(a2>=0.5,1,0)


def main():
    X_std = (X - np.mean(X, axis=0)) / np.std(X, axis=0)
    k=8
    eta=0.1
    max_iter=10000
    W1,b1,W2,b2=train(X_std,y,k,eta,max_iter)
    y_pred=predict(X_std,W1,b1,W2,b2)
    print("Accuracy:",np.mean(y_pred==y))

if __name__=="__main__":
    main()
    

```
